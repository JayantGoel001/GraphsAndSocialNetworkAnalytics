```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_csv("Q2.csv")
```

```python
df.head()
```

```python
G = nx.from_pandas_edgelist(df,create_using=nx.DiGraph())
```

```python
nx.draw_networkx(G,node_color='green',node_size=1000)
```

```python
def generateDegreeDistribution(degree,title):
    degrees = []
    names = []
    for name,deg in degree:
        names.append(name)
        degrees.append(deg)
    plt.plot(names,degrees,color='black')
    plt.xlabel("Nodes")
    plt.ylabel("Degree")
    plt.title(title)
    plt.show()
    
    X = []
    y = []
    
    d = {}
    for i in degrees:
        if i not in d:
            d[i] = 0
        d[i] += 1
    
    for key,value in d.items():
        X.append(key)
        y.append(value/len(degrees))
    
    plt.bar(X,y,color='green')
    plt.xlabel("Degree")
    plt.ylabel("Probability")
    plt.title("Probability Distribution")
    plt.show()
```

```python
print("Probablity Distribution of Degree, In Degree & Out Degree are : ")
```

```python
degree = G.degree()
generateDegreeDistribution(degree,"Degree")
```

```python
degree = G.in_degree()
generateDegreeDistribution(degree,"In Degree")
```

```python
degree = G.out_degree()
generateDegreeDistribution(degree,"Out Degree")
```

```python
print("Average Degree :",sum([y for _,y in nx.degree(G)])/len(G.nodes))
```
