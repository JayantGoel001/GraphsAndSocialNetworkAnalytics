```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_csv("tvshow_edges.csv")
df.head()
```

```python
G = nx.Graph()
G.add_edges_from(df.values.tolist())
```

```python
print("Number of Nodes :",G.number_of_nodes())
print("Number of Edges :",G.number_of_edges())
```

```python
print("Type of Graph is : ")
print("Directed Graph: ",G.is_directed())
print("Multi Graph : ",G.is_multigraph())

print("Graph is Simple undirected Graph.")
```

```python
print("Density of Graph is :",nx.density(G))
```

```python
print("Clustering Coefficient Of Graph is :",nx.clustering(G))
```

```python
degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees,label="Degree")
plt.title("Degree Distribution Graph")
plt.legend()
plt.show()
```

```python
print("Number of Connected Components :",nx.number_connected_components(G))
```

```python
print("Average Shortest path length :",nx.average_shortest_path_length(G))
```

```python
nx.draw(G)
plt.title("Facebook Tv Show Graph")
plt.show()
```
