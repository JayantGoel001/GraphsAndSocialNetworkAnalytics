```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
edge_list = [
    ['A','B'],
    ['A','C'],
    ['A','D'],
    ['B','C'],
    ['B','D'],
    ['B','F'],
    ['C','D'],
    ['C','E'],
    ['C','F'],
    ['D','E'],
    ['D','F'],
    ['E','F'],
    ['F','G'],
    ['F','H'],
    ['F','I'],
    ['F','J'],
    ['G','H'],
    ['G','I'],
    ['H','I'],
    ['H','J'],
    ['I','J']
]
```

```python
G = nx.Graph(edge_list)
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx(G,pos,font_color='white',node_color='black',edge_color='black')
```

```python
# Maximum Clique

print("Maximum Clique :",nx.algorithms.approximation.max_clique(G))
```

```python
# Overlapping Communities Using Clique Percolation Method
communities = list(nx.algorithms.community.k_clique_communities(G,4))
print("Overlapping Communities using Percolation Method for Clique of Size 4 : ")
print(communities)
```

```python
nx.draw_networkx(G,pos,font_color='white')
for i in range(len(communities)):
    r = np.random.randint(10,100)
    g = np.random.randint(10,100)
    b = np.random.randint(10,100)
    nx.draw_networkx_nodes(G,pos,nodelist=list(communities[i]),node_color='#'+str(r) + str(g) + str(b))
```
