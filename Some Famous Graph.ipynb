```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
p = nx.petersen_graph()
nx.draw(p,node_color='cyan',node_size=650,with_labels=True)
```

```python
t = nx.tutte_graph()
nx.draw(t,node_color="#A0CBE2",node_size=600,with_labels=True)
```

```python
s = nx.sedgewick_maze_graph()
nx.draw(s,node_color="#C12345",node_size=700,with_labels=True)
```
