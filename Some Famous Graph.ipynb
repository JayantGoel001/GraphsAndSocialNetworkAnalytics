```python
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
p = nx.petersen_graph()
nx.draw(p,node_color='cyan',node_size=650,with_labels=True)
```

```python
t = nx.tutte_graph()
nx.draw(t,node_color="#A0CBE2",node_size=600,with_labels=True)
```

```python
s = nx.sedgewick_maze_graph()
nx.draw(s,node_color="#C12345",node_size=700,with_labels=True)
```

```python
tt = nx.tetrahedral_graph()
nx.draw(tt,node_color='green',node_size=700,with_labels=True)
```

# Famous Social Network Graph

```python
k = nx.karate_club_graph()
nx.draw(k,node_color="#A0CBE2",node_size=650,with_labels=True)
```

```python
print(k.is_directed())
print(nx.info(k))
```

```python
plt.figure(figsize=(10,7))
dsw = nx.davis_southern_women_graph()
nx.draw(k,node_color='c',node_size=650,with_labels=True)
```

```python
print(dsw.edges(data=True))
```
