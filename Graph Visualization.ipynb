```python
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import display,Image
plt.style.use('seaborn')
```

```python
Image('graph.png')
```

```python
G = nx.Graph()
```

```python
LIGHT_BLUE = '#A0CBE2'
G.add_edges_from([('A','B'),('B','D'),('D','C'),('D','E'),('D','F')])
```

```python
plt.figure(figsize=(4,4))
nx.draw(G,with_labels=True,node_size=800,node_color = LIGHT_BLUE)
plt.show()
```

# A Circular Random Graph

```python
G = nx.cycle_graph(24)
```

```python
pos = nx.spring_layout(G,iterations=190)
```

```python
nx.draw(G,pos,node_color=range(24),node_size=800,cmap=plt.cm.coolwarm)
plt.savefig('cr-graph.png')
plt.show()
```
