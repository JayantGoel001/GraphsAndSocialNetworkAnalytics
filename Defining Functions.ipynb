```python
import matplotlib.pyplot as plt
import networkx as nx
plt.style.use('seaborn')
```

```python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(1,3),(1,4)])
    nx.draw(G,node_size=800,node_color='orange',with_labels=True)
```

```python
draw_graph()
```

```python
# Defining A Function call draw_barabasi_albert
def draw_barabasi_albert():
    G = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(G)
    nx.draw(G,with_labels=True,node_color='#52AF84')
    plt.show()
    plt.hist(hst,bins=20,color='green')
```

```python
draw_barabasi_albert()
```

```python
def simpleGraph():
    G = nx.Graph()
    
    G.add_edge('A','B')
    G.add_edges_from([('B','C'),('C','A')])
    plt.figure(figsize=(5,5))
    nx.draw(G,with_labels=True,node_color='y',node_size=500)
```

```python
simpleGraph()
```

# Deleting A Node

```python
def deleteNodes():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
    plt.figure(figsize=(4,4))
    nx.draw(G,with_labels=True,node_color='r',node_size=500)
    plt.show()
    
    G.remove_node(3)
    plt.figure(figsize=(4,4))
    nx.draw(G,with_labels=True,node_color='r',node_size=500)
    plt.show()
```

```python
deleteNodes()
```

```python
def deleteEdge():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(3,4),(4,1),(2,4),(1,3)])
    plt.figure(figsize=(4,4))
    nx.draw(G,with_labels=True,node_color='b',node_size=500)
    plt.show()
    
    G.remove_edges_from([(2,3),(2,1),(2,4)])
    plt.figure(figsize=(4,4))
    nx.draw(G,with_labels=True,node_color='b',node_size=500)
    plt.show()
```

```python
deleteEdge()
```
