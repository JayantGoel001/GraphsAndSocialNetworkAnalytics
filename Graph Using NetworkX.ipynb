```python
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
def showNetork(G,with_labels=False,node_color='green',node_size=700):
    fig,ax = plt.subplots()
    nx.draw_networkx(G,with_labels=with_labels,node_color=node_color,node_size=node_size,ax=ax)
    ax.tick_params(left=True,bottom=True,labelleft=True,labelbottom=True)
```

```python
#Create A An Empty Graph with No nodes and Edges
G = nx.Graph()
```

```python
#Add A Single Node
G.add_node(1)
```

```python
#Add A List Of Nodes
G.add_nodes_from([2,3])
```

```python
showNetork(G)
```

```python
#Adding Edges to a graph
G.add_edge(1,2)
```

```python
showNetork(G)
```

```python
#Add edges using unpack edge tuples
e = (2,3)
G.add_edge(*e)
```

```python
showNetork(G)
```

```python
#Adding list of edges
G.add_edges_from([(1,2),(1,3)])
```

```python
showNetork(G)
```

```python
#find number of nodes And Edges
print(G.number_of_nodes())
print(G.number_of_edges())
```

```python
#Properties of A Graph
print(G.nodes())
print(G.edges())
```

```python
#Getting The neighbor of a connected graph
print(list(G.neighbors(1)))
```

```python
#The Number Of degrees in a Graph
print(G.degree())
```

```python
#Find If A Graph has an Edge
print(G.has_edge(2,4))
print(G.has_edge(2,3))
```

```python
#Find If A Graph has an Node
print(G.has_node(2))
print(G.has_node(4))
```

```python
#Node Manipulation
H = nx.Graph()
edgeList = [(0,1),(1,2),(2,3),(0,3)]
H.add_edges_from(edgeList)
```

```python
showNetork(H,with_labels=True,node_color='orange')
```

```python
H.remove_node(3)
```

```python
showNetork(H,with_labels=True,node_color='orange')
```

```python
H.remove_nodes_from([1,2])
```

```python
showNetork(H,with_labels=True,node_color='orange')
```

```python
#Adding Attribute to our Graph
B = nx.Graph(weekday='Friday',year='2020',month='September',day=26)

B.graph['day'] = 21
```

```python
B.graph
```

```python
showNetork(B)
```
