```python
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
def showNetork(G):
    fig,ax = plt.subplots()
    nx.draw_networkx(G,node_color='green',node_size=700,ax=ax)
    ax.tick_params(left=True,bottom=True,labelleft=True,labelbottom=True)
```

```python
#Create A An Empty Graph with No nodes and Edges
G = nx.Graph()
```

```python
#Add A Single Node
G.add_node(1)
```

```python
#Add A List Of Nodes
G.add_nodes_from([2,3])
```

```python
showNetork(G)
```

```python
#Adding Edges to a graph
G.add_edge(1,2)
```

```python
showNetork(G)
```

```python
#Add edges using unpack edge tuples
e = (2,3)
G.add_edge(*e)
```

```python
showNetork(G)
```

```python
#Adding list of edges
G.add_edges_from([(1,2),(1,3)])
```

```python
showNetork(G)
```

```python
#find number of nodes And Edges
print(G.number_of_nodes())
print(G.number_of_edges())
```
