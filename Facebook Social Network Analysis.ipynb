```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
fb = nx.read_edgelist('facebook_combined.txt',create_using=nx.Graph(),nodetype=int)
```

```python
pos = nx.spring_layout(fb)

nx.draw_networkx(fb,pos=pos,with_labels=False,node_size=35,node_color='r')
```

```python
print(nx.info(fb))
```

```python
print(nx.betweenness_centrality(fb,normalized=True))
```

```python
print(fb.degree())
```

```python
print(list(fb.neighbors(107)))
```

```python
print(fb.order()) #nodes
print(fb.size())  #edges
```

```python
print(nx.triangles(fb))
```

```python
print(nx.clustering(fb))
```

```python
plt.figure(figsize=(10,7))
hst = nx.degree_histogram(fb)
plt.hist(hst,bins=10,color='orange')
```

```python
bc = nx.betweenness_centrality(fb,normalized=True)
```

```python
bc_values = np.array(list(bc.values()))
```

```python
bc_values = (bc_values-np.mean(bc_values))/np.std(bc_values)
```

```python
plt.bar(range(0,500),list(bc.values())[:500],align='center')
```
