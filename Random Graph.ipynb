# A Random Graph called grid_2d_graph

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
G = nx.grid_2d_graph(4,4)

pos = nx.spring_layout(G,iterations=100)

plt.subplot(221)
nx.draw(G,pos,font_size=8)

plt.subplot(222)
nx.draw(G,pos,node_color='k',with_labels=False,node_size=0)

plt.subplot(223)
nx.draw(G,pos,node_color='g',node_size=1,with_labels=False)

plt.subplot(224)

H = G.to_directed()

nx.draw(H,pos,node_color='b',node_size=20)

plt.savefig('random-grid.png')
plt.show()
```

# Circular Tree

```python
G = nx.star_graph(20)
```

```python
pos = nx.spring_layout(G,iterations=20)
colors = range(20)

nx.draw(G,pos,node_color="#A0CBE2",edge_color=colors,width=4,edge_cmap=plt.cm.Accent,with_labels=False)
```
