```python
import networkx as nx
import matplotlib.pyplot as plt
import os
plt.style.use('seaborn')
```

# Write A Graph in python pickle format to disk

```python
G = nx.path_graph(5)
nx.draw(G,with_labels=True)
```

```python
nx.write_gpickle(G,'path_graph.qpickle')
```

# Read Graph Object from Python Pickle Format

```python
data = nx.read_gpickle('path_graph.qpickle')
nx.draw(data,with_labels=True)
```
