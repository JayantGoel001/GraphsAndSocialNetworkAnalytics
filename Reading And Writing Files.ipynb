```python
import networkx as nx
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Write A Graph in python pickle format to disk

```python
G = nx.path_graph(20)
nx.draw(G,with_labels=True)
```

```python
nx.write_gpickle(G,'path_graph.gpickle')
```

# Read Graph Object from Python Pickle Format

```python
data = nx.read_gpickle('path_graph.gpickle')
nx.draw(data,with_labels=True)
```

```python
print(nx.info(data))
```

# Reading And Writing Edge List

```python
nx.write_edgelist(G,'path_edge.text',data=True)
```

```python
edge_list = nx.read_edgelist('path_edge.text')
nx.draw(edge_list)
```

# Reading files using OPEN python function

```python
file = open('path_edge.text','rb')

G = nx.read_edgelist(file)

nx.draw(G,with_labels=True,node_color="red")
file.close()
```

```python
print(nx.info(G))
```
